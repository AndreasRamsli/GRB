```python
bad_GRB,good_GRB = [],[]
GRBs = {}
for T in T0s:
    dt = datetime.fromisoformat(T)
    if T in good_T0s:
        GRBs['GRB'+datetime.strftime(dt,'%y%m%d')] = {'constant':True, 'theta': None, 'phi': None}
    elif T in bad_T0s:
        GRBs['GRB'+datetime.strftime(dt,'%y%m%d')] = {'constant':False, 'theta': None, 'phi': None}
    else:
        GRBs['GRB'+datetime.strftime(dt,'%y%m%d')] = {'constant':0, 'theta': None, 'phi': None}
GRBs
```

```python
from matplotlib.patches import Patch

thetas = [GRBs[key]['theta'] for key in GRBs.keys()]
phis = [GRBs[key]['phi'] for key in GRBs.keys()]
constant = [GRBs[key]['constant'] for key in GRBs.keys()]
colors = []
for c in constant:
    if isinstance(c,bool):
        if c == True:
            colors.append('green')
        else:
            colors.append('red')
    else:
        colors.append('black')
        
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.scatter(np.deg2rad(phis), np.deg2rad(thetas), color=colors, marker='.',s=200)

ax.set_xticks(np.arange(0, 2*np.pi, np.pi/4), minor=True)
ax.set_xticks(np.deg2rad([0, 45, 90, 135, 180, 225, 270, 315]), minor=False)
ax.set_xticklabels([u'0\u00b0', u'45\u00b0', u'90\u00b0', u'135\u00b0', u'180\u00b0', u'-135\u00b0', u'-90\u00b0', u'-45\u00b0'],c='b')
ax.set_rticks(np.deg2rad([0, 20, 40, 60, 80, 100, 120, 140, 160, 180]))
ax.set_yticklabels([u'0\u00b0 ',u'20\u00b0', u'40\u00b0',u'60\u00b0', u'80\u00b0',u'100\u00b0', u'120\u00b0',u'140\u00b0', u' 160\u00b0',u'180\u00b0'],fontsize=8,c='purple',label='theta')
ax.set_rlabel_position(180)
theta = np.linspace(0, 2*np.pi, 1000)
ax.xaxis.grid(True,color='blue',linestyle='-')
ax.yaxis.grid(True,color='purple',linestyle='-')

ax.fill_between(theta, y1=0, y2=np.deg2rad(60), color='grey', alpha=0.7) #columbus
ax.fill_between(np.deg2rad(np.linspace(45,135,1000)), y1=0, y2=np.deg2rad(180), color='grey', alpha=0.7) #earth
ax.set_ylim([0, 3.15])

ax.text(0.9,2.5,"I",fontsize='x-large')
ax.text(0.3,0.7,"II",fontsize='x-large')
ax.text(-0.01,4,chr(966),fontsize='xx-large',color='blue')
ax.text(0.01,0.25,chr(952),fontsize='xx-large',color='purple')

labels = ['OK', 'Not OK', 'Not Analyzed','I Occulted','II Occulted']
patches = [Patch(color='green', label=labels[0]), 
           Patch(color='red', label=labels[1]), 
           Patch(color='black', label=labels[2]),
           Patch(color='grey', label=labels[3]),
           Patch(color='grey', label=labels[4])]

for i, (x, y) in enumerate(zip(phis, thetas)):
    ax.text(np.deg2rad(x+6.3), np.deg2rad(y-1), str(i+1))

plt.legend(handles=patches, loc='upper left', bbox_to_anchor=(1,1))
plt.title(f'Incomming {chr(952)}, {chr(966)} angle for GRBs',fontsize=20, fontweight='bold')

plt.savefig('polar_plot.png',dpi=300)
plt.show() #numbering in sorted GRB order
```

```python
import plotly.graph_objects as go
import numpy as np

config = {
  'toImageButtonOptions': {
    'format': 'png', # one of png, svg, jpeg, webp
    'filename': '3Dpolar_plot',
    'height': 500,
    'width': 700,
    'scale':6 # Multiply title/legend/axis/canvas sizes by this factor
  }
}
thetas = [GRBs[key]['theta'] for key in GRBs.keys()]
phis = [GRBs[key]['phi'] for key in GRBs.keys()]
constant = [GRBs[key]['constant'] for key in GRBs.keys()]

# Determine color of each point
colors = []
for c in constant:
    if isinstance(c, bool):
        if c == True:
            colors.append('green')
        else:
            colors.append('red')
    else:
        colors.append('black')

# Convert spherical coordinates to Cartesian coordinates
x = np.sin(np.deg2rad(thetas)) * np.cos(np.deg2rad(phis))
y = np.sin(np.deg2rad(thetas)) * np.sin(np.deg2rad(phis))
z = np.cos(np.deg2rad(thetas))

# Create the scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, showlegend=False,mode='markers', marker=dict(color=colors, size=4))])

# Add lines from the origin to each point
fig.add_trace(go.Scatter3d(x=[0, x[0]], y=[0, y[0]], z=[0, z[0]], mode='lines', line=dict(color='black')))
for i in range(len(thetas)):
    fig.add_trace(go.Scatter3d(x=[0, x[i]], y=[0, y[i]], z=[0, z[i]], mode='lines', showlegend=False,line=dict(color='black')))

# Add data point labels
for i in range(len(thetas)):
    fig.add_trace(go.Scatter3d(x=[x[i]], y=[y[i]], z=[z[i]], mode='text', showlegend=False,
                               text=str(i+1), textposition='top center', textfont=dict(size=10, color='black')))

# Add shaded areas to the plot
fig.add_trace(go.Surface(x=[0, 0], y=[-1, 1], z=[0, 0],showlegend=False, autocolorscale=True, opacity=0.7))
fig.add_trace(go.Surface(x=[0, 0], y=[0, 0], z=[-1, 1],showlegend=False, autocolorscale=True, opacity=0.7))

# Add the columbus module surface
theta, phi = np.meshgrid(np.linspace(0, 60, 100), np.linspace(45, 135, 100))
r = np.cos(np.deg2rad(theta))
x = r * np.cos(np.deg2rad(phi))
y = r * np.sin(np.deg2rad(phi))
z = np.sin(np.deg2rad(theta))
fig.add_trace(go.Surface(x=x, y=y, z=z, autocolorscale=True,showlegend=False, opacity=0.7))

# Add a cone to the plot to trace out the region where theta is between 0 and 60 degrees
theta_cone = np.linspace(0, 60, 50)
phi_cone = np.linspace(0, 360, 50)
theta_cone, phi_cone = np.meshgrid(theta_cone, phi_cone)
x_cone = np.sin(np.deg2rad(theta_cone)) * np.cos(np.deg2rad(phi_cone))
y_cone = np.sin(np.deg2rad(theta_cone)) * np.sin(np.deg2rad(phi_cone))
z_cone = np.cos(np.deg2rad(theta_cone))
fig.add_trace(go.Surface(x=x_cone, y=y_cone, z=z_cone, showlegend=False,opacity=0.3))

# Set the axis and layout settings
fig.update_layout(scene=dict(xaxis=dict(showticklabels=False, title='X'),
                             yaxis=dict(showticklabels=False, title='Y'),
                             zaxis=dict(showticklabels=False, title='Z'),
                             camera=dict(up=dict(x=0, y=0, z=-1), eye=dict(x=1.2, y=0, z=0.2))))

#fig.show(config=config)

#fig.write_image("/Users/andreas/phys/master/GRB/cordinates/3D_polarPlot_lookingUp.png",scale=6)
fig.show()
```
