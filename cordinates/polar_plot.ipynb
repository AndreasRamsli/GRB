```python
thetas = [GRBs[key]['theta'] for key in GRBs.keys()]
phis = [GRBs[key]['phi'] for key in GRBs.keys()]
constant = [GRBs[key]['constant'] for key in GRBs.keys()]

# Determine color of each point
colors = []
for c in constant:
    if isinstance(c, bool):
        if c == True:
            colors.append('green')
        else:
            colors.append('red')
    else:
        colors.append('black')

# Convert spherical coordinates to Cartesian coordinates
x = np.sin(np.deg2rad(thetas)) * np.cos(np.deg2rad(phis))
y = np.sin(np.deg2rad(thetas)) * np.sin(np.deg2rad(phis))
z = np.cos(np.deg2rad(thetas))

# Create the scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, showlegend=False,mode='markers', marker=dict(color=colors, size=4))])

# Add lines from the origin to each point
fig.add_trace(go.Scatter3d(x=[0, x[0]], y=[0, y[0]], z=[0, z[0]], mode='lines', line=dict(color='black')))
for i in range(len(thetas)):
    fig.add_trace(go.Scatter3d(x=[0, x[i]], y=[0, y[i]], z=[0, z[i]], mode='lines', showlegend=False,line=dict(color='black')))

# Add data point labels
for i in range(len(thetas)):
    fig.add_trace(go.Scatter3d(x=[x[i]], y=[y[i]], z=[z[i]], mode='text', showlegend=False,
                               text=str(i+1), textposition='top center', textfont=dict(size=10, color='black')))

# Add shaded areas to the plot
fig.add_trace(go.Surface(x=[0, 0], y=[-1, 1], z=[0, 0],showlegend=False, autocolorscale=True, opacity=0.7))
fig.add_trace(go.Surface(x=[0, 0], y=[0, 0], z=[-1, 1],showlegend=False, autocolorscale=True, opacity=0.7))

# Add the columbus module surface
theta, phi = np.meshgrid(np.linspace(0, 60, 100), np.linspace(45, 135, 100))
r = np.cos(np.deg2rad(theta))
x = r * np.cos(np.deg2rad(phi))
y = r * np.sin(np.deg2rad(phi))
z = np.sin(np.deg2rad(theta))
fig.add_trace(go.Surface(x=x, y=y, z=z, autocolorscale=True,showlegend=False, opacity=0.7))

# Add a cone to the plot to trace out the region where theta is between 0 and 60 degrees
theta_cone = np.linspace(60, 0, 50)
phi_cone = np.linspace(0, 360, 50)
theta_cone, phi_cone = np.meshgrid(theta_cone, phi_cone)
x_cone = np.sin(np.deg2rad(theta_cone)) * np.cos(np.deg2rad(phi_cone))
y_cone = np.sin(np.deg2rad(theta_cone)) * np.sin(np.deg2rad(phi_cone))
z_cone = np.cos(np.deg2rad(theta_cone))
fig.add_trace(go.Surface(x=x_cone, y=y_cone, z=z_cone, showlegend=False,opacity=0.3))

# Set the axis and layout settings
fig.update_layout(scene=dict(xaxis=dict(showticklabels=False, title='X'),
                             yaxis=dict(showticklabels=False, title='Y'),
                             zaxis=dict(showticklabels=False, title='Z'),
                             camera=dict(up=dict(x=0, y=0, z=-1), eye=dict(x=1.2, y=0, z=0.2))))

#fig.write_image("/Users/andreas/phys/master/GRB/cordinates/3D_polarPlot_lookingUp.png",scale=6)
fig.show()
```
