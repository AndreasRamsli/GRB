```python
import pandas as pd
# Read the file with multiple spaces as the delimiter
df = pd.read_csv('ISS_theta_phi_angles.txt', delimiter=r"\s+",usecols=[0,1,4,5])
df.columns = ['date', 'time', 'theta', 'phi']

desired_dates = ['180720','181222','190206','190305','190606','200415','200521','200716','201227','210424','210619','211211']

# Extract date from 'date' column and format it
df['date'] = df['date'].str.replace('/', '').str[2:]

# Filter only the rows with the desired dates
df = df[df['date'].isin(desired_dates)]
df = df.reset_index(drop=True)
df.index = df.index + 1
df
```
