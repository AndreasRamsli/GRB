```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
%run /Users/andreas/phys/master/GRB/src/paths/mac.ipynb
```

```python
dir_path = '/Users/andreas/phys/master/GRB/xspec/ASIM/ASIMFits/RSP/drm_120/'
```

```python
os.listdir('drm_120')
```

```python
eff_area = []
filenames = os.listdir('drm_120')
phi_angles = []
for fn in filenames:
    if fn == '.DS_Store':
        continue
    else:
        phi = fn[20:-4]
        HED_DRM = pd.read_csv(dir_path+fn, sep=' ',header=None).T.to_numpy(dtype=np.float32)
        eff_area.append(sum(HED_DRM[17,:]))
        phi_angles.append(phi)
        
phi_angles = np.array(phi_angles,dtype=np.float16)
eff_area = np.array(eff_area,dtype=np.float16)
index_sort = np.argsort(phi_angles)
phi_angles = phi_angles[index_sort]
eff_area = eff_area[index_sort]
```

```python
plt.plot(phi_angles,eff_area)
plt.xlabel(chr(966)+'-angle deg')
plt.ylabel('Sum of effective area per bin')
plt.title(chr(952)+' = 120 deg, E = 0.5 MeV')
plt.grid()
plt.show()
```
